**6. Outlier Detection:**
* Check for the outliers by using the IQR method.

```python
import numpy as np

# Selecting only the numerical features
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Calculating Q1 and Q3 for each numerical feature
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)

# Calculating IQR for each numerical feature
IQR = Q3 - Q1

# Defining upper and lower bounds for outlier detection
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identifying outliers for each numerical feature
outliers = ((numerical_df < lower_bound) | (numerical_df > upper_bound))

# Printing the outliers for each feature
for column in outliers.columns:
  if outliers[column].any():
    print(f"Outliers in {column}:")
    print(numerical_df[outliers[column]][column])
```

**7. Conditional Probabilities:**
* What percent of customers have purchased KP281, KP481, or KP781?

* Create frequency tables and calculate the percentage as follows
####  Product – Gender
1.  Percentage of a Male customer purchasing a treadmill
2. Percentage of a Female customer purchasing KP781 treadmill
3. Probability of a customer being a Female given that Product is
KP281

####Product – Age
1. Percentage of customers with Age between 20s and 30s among all
customers

####Product – Income
1. Percentage of a low-income customer purchasing a treadmill
2. Percentage of a high-income customer purchasing KP781
treadmill
3. Percentage of customer with high-income salary buying treadmill
given that Product is KP781

#### Product – Fitness
1. Percentage of customers that have fitness level 5
2. Percentage of a customer with Fitness Level 5 purchasing KP781
treadmill
3. Percentage of customer with fitness level 5 buying KP781
treadmill

####Product - Marital Status
1.  Percentage of a customers who are partnered using treadmills

```python
# Conditional Probabilities:What percent of customers have purchased KP281, KP481, or KP781?

# Calculating the total number of customers
total_customers = len(df)

# Calculating the number of customers who purchased each product
kp281_customers = df[df['Product'] == 'KP281'].shape[0]
kp481_customers = df[df['Product'] == 'KP481'].shape[0]
kp781_customers = df[df['Product'] == 'KP781'].shape[0]

# Calculating the percentage of customers who purchased each product
kp281_percent = (kp281_customers / total_customers) * 100
kp481_percent = (kp481_customers / total_customers) * 100
kp781_percent = (kp781_customers / total_customers) * 100

# Print the results
print(f"Percentage of customers who purchased KP281: {kp281_percent:.2f}%")
print(f"Percentage of customers who purchased KP481: {kp481_percent:.2f}%")
print(f"Percentage of customers who purchased KP781: {kp781_percent:.2f}%")
```

```python
# product -gender

# Creating a frequency table for Product and Gender
product_gender_table = pd.crosstab(df['Product'], df['Gender'])

# 1. Percentage of a Male customer purchasing a treadmill
male_customers = product_gender_table['Male'].sum()
total_customers = product_gender_table.sum().sum()
male_purchase_percent = (male_customers / total_customers) * 100

# 2. Percentage of a Female customer purchasing KP781 treadmill
female_kp781 = product_gender_table.loc['KP781', 'Female']
female_customers = product_gender_table['Female'].sum()
female_kp781_percent = (female_kp781 / female_customers) * 100

# 3. Probability of a customer being a Female given that Product is KP281
female_kp281 = product_gender_table.loc['KP281', 'Female']
kp281_customers = product_gender_table.loc['KP281'].sum()
prob_female_given_kp281 = (female_kp281 / kp281_customers) * 100

# Print the results
print(f"1. Percentage of Male customers purchasing a treadmill: {male_purchase_percent:.2f}%")
print(f"2. Percentage of Female customers purchasing KP781: {female_kp781_percent:.2f}%")
print(f"3. Probability of a customer being Female given KP281: {prob_female_given_kp281:.2f}%")
```

```python
# gender- age

# Creating an age group column
df['Age_Group'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70],
                          labels=['20s', '30s', '40s', '50s', '60s'],
                          right=False)

# Creating a frequency table for Product and Age Group
product_age_table = pd.crosstab(df['Product'], df['Age_Group'])

# Calculating the total number of customers
total_customers = product_age_table.sum().sum()

# Calculating the number of customers in their 20s and 30s
customers_20s_30s = product_age_table['20s'].sum() + product_age_table['30s'].sum()

# Calculating the percentage
percent_20s_30s = (customers_20s_30s / total_customers) * 100

# Print the result
print(f"Percentage of customers with Age between 20s and 30s: {percent_20s_30s:.2f}%")
```

```python
# product- income

# Defining the low and high income thresholds
low_income_threshold = 50000
high_income_threshold = 70000

# Creating income group column
df['Income_Group'] = pd.cut(df['Income'], bins=[0, low_income_threshold, high_income_threshold, np.inf],
                            labels=['Low', 'Medium', 'High'], right=False)

# Creating the frequency table for Product and Income Group
product_income_table = pd.crosstab(df['Product'], df['Income_Group'])

# 1. Percentage of low-income customers purchasing a treadmill
low_income_customers = product_income_table['Low'].sum()
total_customers = product_income_table.sum().sum()
low_income_percent = (low_income_customers / total_customers) * 100

# 2. Percentage of high-income customers purchasing KP781
high_income_kp781 = product_income_table.loc['KP781', 'High']
high_income_customers = product_income_table['High'].sum()
high_income_kp781_percent = (high_income_kp781 / high_income_customers) * 100

# 3. Percentage of customers with high-income buying treadmill given that Product is KP781
kp781_customers = product_income_table.loc['KP781'].sum()
high_income_given_kp781_percent = (high_income_kp781 / kp781_customers) * 100

# Print the results
print(f"1. Percentage of low-income customers purchasing a treadmill: {low_income_percent:.2f}%")
print(f"2. Percentage of high-income customers purchasing KP781: {high_income_kp781_percent:.2f}%")
print(f"3. Percentage of high-income customers buying treadmill given KP781: {high_income_given_kp781_percent:.2f}%")
```

```python
# product - fitness

# Creating frequency table for Product and Fitness
product_fitness_table = pd.crosstab(df['Product'], df['Fitness'])

# 1. Percentage of customers that have fitness level 5
fitness_level_5_customers = product_fitness_table[5].sum()
total_customers = product_fitness_table.sum().sum()
fitness_level_5_percent = (fitness_level_5_customers / total_customers) * 100

# 2. Percentage of customers with Fitness Level 5 purchasing KP781 treadmill
kp781_fitness_level_5 = product_fitness_table.loc['KP781', 5]
fitness_level_5_customers = product_fitness_table[5].sum()
kp781_fitness_level_5_percent = (kp781_fitness_level_5 / fitness_level_5_customers) * 100

# Print the results
print(f"1. Percentage of customers with fitness level 5: {fitness_level_5_percent:.2f}%")
print(f"2. Percentage of customers with Fitness Level 5 purchasing KP781: {kp781_fitness_level_5_percent:.2f}%")
```

```python
# product - marital status

# Createing the frequency table for Product and MaritalStatus
product_marital_table = pd.crosstab(df['Product'], df['MaritalStatus'])

# Calculating the number of partnered customers using treadmills
partnered_customers = product_marital_table['Partnered'].sum()

# Calculating the total number of customers
total_customers = product_marital_table.sum().sum()

# Calculating the percentage
partnered_percent = (partnered_customers / total_customers) * 100

# Print the result
print(f"Percentage of partnered customers using treadmills: {partnered_percent:.2f}%")
```
